## Separa X de Y

```python
X = dt_model.drop(columns=["riesgo"])
y = dt_model["riesgo"]
```

# Escalado de datos

Por qué se hace?

Variables están en escalas diferentes

Ayuda mucho a modelos como regresión logística

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# En esta parte se usa "fit_transform" porque es la prueba del modelo.
En el entrenamiento debe ser "transform"

```python
import joblib
import os

# Create the 'models' directory if it doesn't exist
os.makedirs("models", exist_ok=True)

joblib.dump(scaler, "models/scaler.pkl")
```

# División Train / Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

¿Por qué stratify=y?

Mantiene la proporción de estudiantes en riesgo

Evita sesgos
